```python
from bs4 import BeautifulSoup # Importing the BeautifulSoup package
import requests # Importing requests library
import csv
import pandas as pd
from time import sleep
from random import randint
```

```python
#here, enter the number of ages to be scrapped
# new - https://www.newegg.com/p/pl?d=laptop&N=4841%2050010772%204814&PageSize=96&page=
#old - https://www.newegg.com/p/pl?d=laptop&N=4841%2050010772%204016&PageSize=96&page=

lap_nm=[]
lap_pr=[]
lap_url=[]
page_num=input("Enter Number of page")
for i in range (1,int(page_num)+1):
    url=("https://www.newegg.com/p/pl?d=laptop&N=4841%2050010772%204016&PageSize=96&page=")+str(i) # Requesting the content of the URL
    req = requests.get(url)
    content=BeautifulSoup(req.content,'html.parser')
    name=content.find_all('a' ,{'class':'item-title'})
    price = content.find_all('li',{'class':'price-current'})
    url = content.find_all('a',{'class':'item-title'})
    sleep(randint(4,10))
    print("Laptop in page"+str(i))
    print(len(name))
    
    for i in name:
        lap_nm.append(i.text)
    for i in price:
        lap_pr.append(i.text)
    for i in url:
        lap_url.append(i['href'])
```

```python
for i in lap_pr:
    print(i)
```
